```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
years = [2015, 2016, 2017, 2018, 2019]
height = [110, 115, 120, 125, 132]
fig, ax = plt.subplots()
ax.plot(years, height, color="#00a6ed")
ax.set_xticks(years) 
plt.title("child height along the years")
plt.xlabel('years')
plt.ylabel('height (cm)')
ax.set_xlim(2015,2019)
ax.set_ylim(110.01,132.01)
plt.show()
```

```python
months = ["Jan", "Feb", "Mar", "Apr", "May"]
temperature = [12, 14, 17, 21, 25]
rain = [80, 60, 55, 40, 20]
fig, axs = plt.subplots(1, 2,figsize=(10, 4))

axs[0].plot(months ,temperature,color='#00d269')
axs[0].set_title("temperature along the years")
axs[1].bar(months,rain,color='#00d269')
plt.title("rain amount along the years")
plt.tight_layout() 
plt.show()
```

```python
grades = [70, 85, 90, 76, 65, 95, 88, 73, 80, 78, 67, 92, 60, 83, 89, 55, 91, 74, 69, 87, 81, 79, 62, 77, 93]
fig, axs = plt.subplots(1,2,figsize=(10, 4))
n_bins10 = 10
n_bins5 = 5
axs[0].hist(grades,n_bins10,color='#f8312e')
axs[0].set_title("grades splite in 10 groups")
axs[1].hist(grades,n_bins5,color='#f8312e')
axs[1].set_title("grades splite in 5 groups")
plt.tight_layout() 
plt.show()
```

```python
age_groups = ['10-12', '13-15', '16-18', '19-21']
students_count = [30, 45, 25, 10]
fig, ax = plt.subplots()
ax.bar(age_groups, students_count, color="#ffb02f")
ax.set_xticks(age_groups) 
plt.title("child height along the years")
plt.xlabel('age groups')
plt.ylabel('student count')
plt.show()
```
