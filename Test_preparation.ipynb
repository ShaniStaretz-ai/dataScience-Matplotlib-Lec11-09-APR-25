```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import Markdown,display
def printmd(message,data):
    text="**<u>"+message+"</u>**"
    display(Markdown(text),data)
```

```python
students_df = pd.read_csv("students.csv")
courses_df = pd.read_csv("courses.csv")
teachers_df = pd.read_csv("teachers.csv")
studnet_course_df = pd.read_csv("studnet_course.csv")
students_df.info()
courses_df.info()
teachers_df.info()
studnet_course_df.info()
```

```python
#clean data
students_df.dropna(inplace=True)
students_df.drop_duplicates(inplace=True)

courses_df.dropna(inplace=True)
courses_df.drop_duplicates(inplace=True)

teachers_df.dropna(inplace=True)
teachers_df.drop_duplicates(inplace=True)

studnet_course_df.dropna(inplace=True)
studnet_course_df.drop_duplicates(inplace=True)
```

```python
genders=students_df['gender'].value_counts()
genders_counts=genders.values
genders_counts
genders=list(genders.index)
fig, ax = plt.subplots()
ax.bar(genders, genders_counts, color="#00a6ed")
plt.title("students gender counts")
plt.xlabel('gender')
plt.ylabel('count')
plt.show()
```

```python
ages=students_df['age']
fig, ax = plt.subplots()
ax.hist(ages,'auto',color='#f8312e')
plt.title("Ages splite into groups, automatically")
plt.xlabel('Age')
plt.ylabel('Number of Students')
plt.show()
```

```python
courses_count=courses_df['course_category'].value_counts()
fig, ax = plt.subplots()
ax.bar(courses_count.index,courses_count.values,color='#00d269')
plt.title("courses in each category")
plt.xlabel('category')
plt.ylabel('Number of course')
plt.xticks(rotation=45)
plt.tight_layout() 
plt.show()
```

```python
printmd("the city with the most student:",students_df['city'].value_counts().idxmax())
printmd("the city with the least student:",students_df['city'].value_counts().idxmin())
```

```python
names=courses_df.groupby('course_category')['course_name'].unique()
course_category=list(names.index)
for cat in course_category:
    print(f"In course category '{cat}', the unique course names are: {','.join(['"' + name + '"' for name in list(names[cat])])}.")
```
