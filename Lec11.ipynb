```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('matplotlib_subplot_class_exc.csv')
df
canada_data=df[df['Country']=='Canada']
x_canada_data=canada_data['Year']
y_canada_data=canada_data['GDP per Capita']

y_histogram_data=df['GDP per Capita']

bar_countries=df['Country']
pd.set_option('display.max_rows',None)
y_histogram_data.value_counts().sort_index(ascending=False)
```

```python
fig,ax=plt.subplots(2,2,figsize=(10,10))

ax[0,0].plot(x_canada_data,y_canada_data)
ax[0,0].set_title('GDP per Capita over the years in Canada Country')
ax[0,0].set_xlabel('Years')
ax[0,0].set_ylabel('GDP per Capita')

ax[0,1].hist(y_histogram_data,bins=20, color='r',alpha=1)
ax[0,1].set_title('GDP per Capita ')
ax[0,1].set_xlabel('GDP per Capita')
ax[0,1].set_ylabel('GDP ')

plt.tight_layout()
plt.show()
```
